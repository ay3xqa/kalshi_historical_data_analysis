```python
import pandas as pd

# Convert y_values into a dictionary mapping each day to its outcome price
y_dict = {y[0].date(): y[1] for y in y_values}

# Convert timestamps in raw_df to datetime
raw_df['time'] = pd.to_datetime(raw_df['time'])

outcomes = []
for _, row in raw_df.iterrows():
    row_time = row['time']
    
    # Find the date that should be used for lookup
    lookup_date = (row_time + pd.Timedelta(hours=7)).date()  # Shift time to align with 17:01 EST cutoff

    if lookup_date in y_dict:
        outcome = 'Yes' if row['strike_price'] <= y_dict[lookup_date] else 'No'
    else:
        outcome = 'No'  # Default if no match found

    outcomes.append(outcome)

df = raw_df.copy()
df['outcome'] = outcomes
df
```
